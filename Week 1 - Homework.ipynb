```python
## Question 1 >>> F1-Score Function

def calc_f1_score(tp, fp, fn):
    if not isinstance(tp, int) or not isinstance(fp, int) or not isinstance(fn, int):
        if not isinstance(tp, int):
            print("tp must be int")
        if not isinstance(fp, int):
            print("fp must be int")
        if not isinstance(fn, int):
            print("fn must be int")
        return
    
    if tp <= 0 or fp <= 0 or fn <= 0:
        print("tp, fp, fn must be greater than zero ")
        return
    
    precision = tp/(tp + fp)
    recall = tp/ (tp + fn)
    f1_score = 2 * (precision * recall)/ (precision + recall)

    print(f"Precision value: {precision}")
    print(f"Recall value: {recall}")
    print(f"F1-Score value: {f1_score}")


calc_f1_score(2,3,5)






```

```python
## Question 2 >>> Activation Function

import math

def calc_sigmoid_func(x):
    result = 1 / (1 + math.e **(-x))
    return result

def calc_relu_func(x):

    if x < 0:
        result = 0
    else:
        result = x
    return result

def calc_elu_func(x):
    alpha = 0.01

    if x < 0:
        result = alpha * ((math.e ** x) - 1)
    else:
        result = x
    return result

def calc_activation_func(x, act_func_name):

    if act_func_name == "sigmoid":
        result = calc_sigmoid_func(x)
        print(f"Result: {act_func_name}({x}) = {result}")
    if act_func_name == "relu":
        result = calc_relu_func(x)
        print(f"Result: {act_func_name}({x}) = {result}")
    if act_func_name == "elu":
        result = calc_elu_func(x)
        print(f"Result: {act_func_name}({x}) = {result}")       

def is_number(input):
    try:
        float(input)
        return True
    except ValueError:
        return False
    
def check_user_input_q2():
    x = input("input x =")
    if is_number(x) == False:
        print("x must be a number")
    else:
        x = float(x)
        act_func_name = input("Input Active Function Name (sigmoid | relu | elu)")
        if act_func_name == "sigmoid" or act_func_name == "relu" or act_func_name == "elu" :
            calc_activation_func(x, act_func_name)
            
        else:   
           print("This Activation Function is not Support")
        

check_user_input_q2()
```

```python
## Question 3 >>> Loss Function


import random

def calc_mae (number_sample):
    total_error = 0
    for i in range(number_sample):
        y = random.uniform(0,10)
        y_hat = random.uniform(0,10)
        absolute_error = abs(y_hat - y)
        total_error = total_error + absolute_error

    mae = total_error/number_sample

    return mae

def calc_mse (number_sample):
    total_error = 0
    for i in range(number_sample):
        y = random.uniform(0,10)
        y_hat = random.uniform(0,10)
        absolute_error = (y_hat - y)**2
        total_error = total_error + absolute_error

    mse = total_error/2
    return mse

def is_integer(variable):
    return isinstance(variable, int)


def check_user_input_q3():
    number_sample = int(input("Input number of sample"))
    check_int = is_integer(number_sample)
    if check_int == False:
        print("number of sample must be integer")
        return
    
    else:
        loss_func_name = input("Input loss function name (mae, mse)  = ")
        if loss_func_name == "mae":
            result = calc_mae(number_sample)
            print(f"loss_name:{loss_func_name} | number sample:{number_sample} | loss is {result}")
        elif loss_func_name == "mse":
            result = calc_mse(number_sample)
            print(f"loss_name:{loss_func_name} | number sample:{number_sample} | loss is {result}")

        else:
            print("This loss function is not support") 
            return


            

   
check_user_input_q3()

```

```python
## Question 4 >>> Four Function Calculate Sin(x), Cos(x), Sinh(x), Cosh(x)

import math 

def factorial_fcn(x):
    res = 1
    for i in range(x):
        res *= (i+1)
    return res

def is_radian(number):
    if is_number(number) == True and number >= 0 and number <= 2*math.pi:
        return True
    else:
        return False    
    
def is_positive_integer(var):
    return isinstance(var, int) and var > 0

def is_number(input):
    try:
        float(input)
        return True
    except ValueError:
        return False
    
def approx_sin (x, n):
    sin_approx = 0
    for i in range (n+1):
        coef = (-1)**i
        num = x ** (2*i+1)
        denom = factorial_fcn(2*i+1)
        sin_approx += (coef) * (num/denom)
    return sin_approx

def approx_cos(x,n):
    cos_approx = 0 
    for i in range(n + 1):
        coef = (-1)**i
        num = x**(2*i)
        demon = factorial_fcn(2*i)
        cos_approx += coef * (num/demon)
    
    return cos_approx

def approx_sinh (x, n):
    sinh_approx = 0
    for i in range (n+1):
        coef = 1
        num = x ** (2*i+1)
        denom = factorial_fcn(2*i+1)
        sinh_approx += (coef) * (num/denom)
    return sinh_approx

def approx_cosh (x,n):
    cosh_approx = 0 
    for i in range(n + 1):
        coef = 1
        num = x**(2*i)
        demon = factorial_fcn(2*i)
        cosh_approx += coef * (num/demon)
    
    return cosh_approx

def check_user_input_q4():
    x = input("Input x:")
    if is_number(x) == True:
        x = float(x)
        if is_radian(x) == True:
            n = input("Input n:")
            if is_number(n) == True: 
                n = int(n)
                if is_positive_integer(n) == True:
                    print(f"approx_sin x={x},n={n} >>>> {approx_sin(x,n)}")
                    print(f"approx_cos x={x},n={n} >>>> {approx_cos(x,n)}")
                    print(f"approx_sinh x={x},n={n} >>>> {approx_sinh(x,n)}")
                    print(f"approx_cosh x={x},n={n} >>>> {approx_cosh(x,n)}")
                    return
                    
                else:
                    print("n must be integer type and greater than zero")
                    return
            else:
                print("n must be a number")

        else:
            print("x must be in radian")
            return

    else:
        print("x must be a number")
        return
         
   
check_user_input_q4()


```

```python
def md_nre_single_sample (y , y_hat , n , p) :
    y_root = y ** (1/ n )
    y_hat_root = y_hat ** (1/ n )
    difference = y_root - y_hat_root
    loss = difference ** p
    return loss

md_nre_single_sample(100,99.5,2,1)
```
